```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop unnamed 
```

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
#change weekly study hours 
```

```python
df.head()
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-10","5-10")
df.head()
```

```python
#gender distribution
```

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above data we have analysed that the number of females in the data is more than the number of males
```

```python
#effect of parent education on student scores
```

```python
gb = df.groupby("ParentEduc").agg({
    "MathScore": "mean",
    "ReadingScore": "mean",
    "WritingScore": "mean"
})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot=True)
plt.title("Impact of Parent's Education on Student's scores")
plt.show()
```

```python
#from the above chart we have concluded that parent education have a good impact on student scores
```

```python
#marital status- parents
```

```python
gb = df.groupby("ParentMaritalStatus").agg({
    "MathScore": "mean",
    "ReadingScore": "mean",
    "WritingScore": "mean"
})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot=True)
plt.title("Impact of Parent's Marital Status on Student's scores")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact of Parent Marital status on student scores
```

```python
#distribution of ethnic groups
```

```python
groupA = df.loc[(df['EthnicGroup']=="group A")].count()
groupB = df.loc[(df['EthnicGroup']=="group B")].count()
groupC = df.loc[(df['EthnicGroup']=="group C")].count()
groupD = df.loc[(df['EthnicGroup']=="group D")].count()
groupE = df.loc[(df['EthnicGroup']=="group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show
```
